```python
#Importing Packages 

import cv2 
import numpy as np
import matplotlib.pyplot as plt
cv2.__version__
```

```python
# loading images , reshaping images

img1 = cv2.imread('C:/Users/kajan/OneDrive/Documents/Data Science/forest.jpg')
img2 = cv2.imread('C:/Users/kajan/OneDrive/Documents/Data Science/people-walking.jpg')

type(img1)
print('showing the size of image 1' ,img1.shape)
print('showing the size of image 2' ,img2.shape)


img1_reshaped = cv2.resize(img1,(300,300),interpolation=cv2.INTER_AREA)
img2_reshaped = cv2.resize(img2,(300,300),interpolation=cv2.INTER_AREA)

print('Showing new size for the img2', img2_reshaped.shape)
```

```python
#printing the images

plt.figure(figsize=(10,10))
plt.subplot(121)
plt.imshow(img1_reshaped)
plt.subplot(122)
plt.imshow(img2_reshaped)
```

```python
#Printing RGB Channels funcition is (gca)

plt.figure(figsize=(10,10))
plt.subplot(131)
plt.imshow(img1_reshaped[:,:,0])
plt.gca().set_title('Red Channel')

plt.subplot(132)
plt.imshow(img1_reshaped[:,:,1])
plt.gca().set_title('Green Channel')

plt.subplot(133)
plt.imshow(img1_reshaped[:,:,2])
plt.gca().set_title('Blue Channel')
```

```python
#Adding images to each other 

dst = cv2.addWeighted(img2_reshaped,0.5,img1_reshaped,0.3,0.1)
plt.figure(figsize=(10,10))
plt.imshow(dst)
```

```python
dst_subt = cv2.subtract(img1_reshaped, img2_reshaped)
plt.figure(figsize=(10,10))
plt.imshow(dst_subt)
```

```python
# Working with the thir image. (Image colors range from 0-255), (Use gca() to set titles)

img3 = cv2.imread('C:/Users/kajan/OneDrive/Documents/Data Science/cat.jpg')
print('showing th esize of the images' ,img3.shape)


negative = 255 - img3
plt.figure(figsize=(10,10))
plt.subplot(121)
plt.imshow(img3)
plt.gca().set_title('Original Image')
plt.subplot(122)
plt.imshow(negative)
plt.gca().set_title('Negative Image')
```

```python
#Blurring images

img_blur = cv2.blur(img3,(80,80))
img_filter = cv2.boxFilter(img3,-5,(30,30),normalize=True)
img_gaussianfilter = cv2.GaussianBlur(img3,(41,41),cv2.BORDER_DEFAULT)

plt.figure(figsize=(20,20))
plt.subplot(141)
plt.imshow(img3)
plt.gca().set_title('Original Image')
plt.subplot(142)
plt.imshow(img_blur)
plt.gca().set_title('Blue Image')
plt.subplot(143)
plt.imshow(img_filter)
plt.gca().set_title('Filter Image')
plt.subplot(144)
plt.imshow(img_gaussianfilter)
plt.gca().set_title('Gaussian FIlter Image')
```

```python
#Defining and image 

img_cat_hsv = cv2.cvtColor(img3,cv2.cv2.COLOR_RGB2HSV)
plt.imshow(img_cat_hsv)
```
